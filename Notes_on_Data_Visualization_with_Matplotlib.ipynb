# 1. Basics of Matplotlib

**Importing Matplotlib**

```
import matplotlib.pyplot as plt
```

````python
import matplotlib.pyplot as plt
import numpy as np

#! To display plolts

%matplotlib inline 

#! Or

plt.show()
````

**A Simple Starter Plot**

````python
x_data = [1,2,6,8,10,12,14]
y_data = np.exp(x_data)

plt.plot(x_data, y_data)
plt.title('This is a title')
plt.xlabel('X_data, normally X axis title')
plt.ylabel('Y_data, normally Y axis title')
plt.show()
````



`x_data` and `y_data` can be lists, NumPy, or a Pandas dataframes.


**Subplots**

`plt.subplot` will show multiple plots on same figure.

````python
#! plt.subplots(number of rows, number of columns, plot number)

#! Create emply figure with multiple plots

plt.subplot(1,4,1)
plt.plot(x_data, y_data)

plt.subplot(1,4,2)
plt.plot(x_data, np.sin(y_data))

plt.subplot(1,4,3)
plt.plot(x_data, np.cos(y_data))

plt.tight_layout()
plt.show()
````

**Figure Size**

We can use `plt.figure(figsize=(width, height)` to adjust the figure size to your desied size.


````python
plt.figure(figsize=(8,4))

plt.subplot(1,4,1)
plt.plot(x_data, y_data)

plt.subplot(1,4,2)
plt.plot(x_data, np.sin(y_data))

plt.subplot(1,4,3)
plt.plot(x_data, np.cos(y_data))

plt.tight_layout()
plt.show()
````

We can also specify the dot per inch (DPI)

````python
plt.figure(figsize=(6,4), dpi=150)

plt.subplot(1,4,1)
plt.plot(x_data, y_data)

plt.subplot(1,4,2)
plt.plot(x_data, np.sin(y_data))

plt.subplot(1,4,3)
plt.plot(x_data, np.cos(y_data))
plt.tight_layout()
plt.show()
````

**Figure Title, Axis Labels and Legends**

````python
#! A title can be added to figure with plt.title('xyz')
#! A Label can be added to figure with plt.xlable('xyz')
#! A legend can be added to figure with plt.legend(loc='upper right')

figure = plt.figure(figsize=(6,4))

x = np.linspace(0, 20, 40)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.plot(x, y1, label = 'Y1')
plt.plot(x, y2, label = 'Y2')
plt.plot(x, y3, label = 'Y3')
plt.title('This is a title')
plt.legend(loc='upper right')
plt.xlabel('This is X label')
plt.ylabel('This is Y label')
plt.show()
````

**Line Colors and Styles**

We can add the color to the line plots using `color` parameter, `color` codes(RGBCYMK), hex codes, and grayscale level.

Color codes (RGBCYMK):
```
'b' = blue
'g' = green
'r' = red
'c' = cyan
'm' = magenta
'y' = yellow
'k' = black
```

Hex Codes:

Example:  #00FF00 is green. #ff0000 is red.
 
Google for any hex code color.






````python
figure = plt.figure(figsize=(6,4))

plt.plot(x, y1, label = 'Y1', color='r') # For red color
plt.plot(x, y2, label = 'Y2', color='yellow') # For yellow color
plt.plot(x, y3, label = 'Y3', color='#00FF00') # For green color
plt.plot(x, y3*2, label = 'Y4', color='0.6') # For grayscale or #B2BEB5
plt.title('This is a title')
plt.legend(loc='upper right')
plt.xlabel('This is X label')
plt.ylabel('This is Y label')
plt.show()
````

**Line Styles and Markers**

We can modify the line styles and markers.



```
# Line Styles

'-' = solid line style
'--' = dashed line style
'-.' = dashed-dot line style
':' = dotted line style
'or' = red circles
'-g' = green solid line
'--' = dashed line with default color
'b' = blue markers with default shape
'^k' = black triangle_up markers connected by a dotted line
```


```
# Line Markers

'.' = point marker
',' = pixel marker
'o' = circle marker
'^' = tirangle_up marker
'v' = triangle_down marker
'<' = triangle_left marker
'>' = triangle_right marker
'1' = tri_down marker
'2' = tri_up marker
'3' = tri_left marker
'4' = tri_right marker
's' = square marker
'p' = pentagon marker
'h' = hexagon1 marker
'H' = hexagon2 marker
'D' = .diamond marker
'd' = thin_diamond marker
'*' = star marker
'+' = plus marker
'x' = x marker
'|' = vline marker
'_' = hline marker

```

````python
#! Line Styles, Line width and Marker Size

figure = plt.figure(figsize=(8,6))

plt.plot(x, y1, label = 'Y1', color='r', ls='solid', marker='*', lw=2, markersize=4)
plt.plot(x, y2, label = 'Y2', color='yellow', ls='dotted', marker='d', lw=2, markersize=4)
plt.plot(x, y3, label = 'Y3', color='#00FF00', ls='dashed', marker='x', lw=2, markersize=4)
plt.plot(x, y3*2, label = 'Y4', color='0.6', ls=':', marker='+', lw=4, markersize=2)
plt.title('This is a title')
plt.legend(loc='upper right')
plt.xlabel('This is X label')
plt.ylabel('This is Y label')
plt.show()
````

**Text and Annotation**

Matplotlib can allow us to add texts or annotation in visualization.

````python
figure = plt.figure(figsize=(8,6))

x_data = [1,2,6,8,10,12,14]
y_data = np.square(x_data)

plt.plot(x_data, y_data)

#! Add the text to the plot
#! plt.text(x position, y position, string, color or style)

plt.text(2,180, 'This text is added at the plot where x = 2, y = 180', color='blue')

plt.title('This is a title')
plt.xlabel('X_data, normally X axis title')
plt.ylabel('Y_data, normally Y axis title')
plt.show()
````

````python
#! Text Annotation in Matplotlib

figure = plt.figure(figsize=(8,6))

x_data = [1,2,6,8,10,12,14]
y_data = np.square(x_data)

plt.plot(x_data, y_data)


plt.annotate('This is the Minimum point', 
             xy=(1,2), 
             xytext=(1.5, 45), arrowprops=dict(facecolor='green'))

plt.annotate('This is the Maximum point', #! string to annotate
             xy=(14,200), #! the position of x and y
             xytext=(2, 185), #! the xy position to add the texts
             bbox=dict(boxstyle='square', fc='yellow'), #! or boxstyle can be 'square'
             arrowprops=dict(arrowstyle='fancy',connectionstyle='angle3, angleA=0, angleB=-90'))

plt.annotate('Middle, estimated', #! string to annotate
             xy=(10,100), #! the position of x and y
             xytext=(5, 100), #! the xy position to add the texts
             bbox=dict(boxstyle='square', fc='none'), # or boxstyle can be 'square'
             arrowprops=dict(arrowstyle='wedge, tail_width=0.4', alpha=0.4))


plt.title('This is a title')
plt.xlabel('X_data, normally X axis title')
plt.ylabel('Y_data, normally Y axis title')
plt.show()
````

More about `plt.text()` and `plt.annotate()` can be found at [Matplotlib Documentation](https://matplotlib.org/).
